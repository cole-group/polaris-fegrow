```python
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
smiles = {'33' : 'C[C@H](NC(=O)COS(N)(=O)=O)c1cc(Cl)cc(-c2ccc(S(N)(=O)=O)cc2)c1',
         '94': 'CNS(=O)(=O)OCC(=O)N[C@@H](C)c1cc(Cl)cc(-c2ccc(S(N)(=O)=O)cc2)c1',
         '22': 'O=C1N[C@@]2(CCOc3ccncc32)C(=O)N1c1cncc2ccccc12',
         '42': 'Cn1ncc2cncc(NC(=O)[C@@H]3CNS(=O)(=O)c4ccc(Cl)cc43)c21'}
```

```python
index = '42'

mol = Chem.MolFromSmiles(smiles[index])
molh = Chem.AddHs(mol)


AllChem.EmbedMolecule(molh)
molh
```

```python
with Chem.SDWriter('in-{0}.sdf'.format(index)) as w:
    w.write(molh)
```
