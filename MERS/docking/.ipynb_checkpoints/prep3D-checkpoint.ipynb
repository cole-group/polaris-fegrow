```python
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
smiles = {'20' : 'CCC(=O)N1CCCc2ccc(N(C(=O)Cc3cncc4ccccc34)N(C)c3cccc(Cl)c3)cc21',
          '41' : 'C[C@H]1CN(c2cncc3ccccc23)C(=O)[C@@]12CN(CCN1CCS(=O)(=O)CC1)C(=O)c1ccc(Cl)cc12',
          '77' : 'CC(C)(C#N)c1ccc(N(Cc2cccc(Cl)c2)C(=O)Cc2cncc3ccccc23)cc1',
          '78' : 'COC(=O)N[C@H](C(=O)N1C[C@H](C(F)(F)F)C[C@H]1C(=O)N[C@H](C[NH3+])C[C@@H]1CCNC1=O)C(C)(C)C',
         }
```

```python
index = '41'

mol = Chem.MolFromSmiles(smiles[index])
molh = Chem.AddHs(mol)


AllChem.EmbedMolecule(molh)
molh
```

```python
with Chem.SDWriter('in-{0}.sdf'.format(index)) as w:
    w.write(molh)
```
